```python
import numpy as np
```

Метод Грама - Ммидта получения ортонормированного базиса:

```python
def gram_schmidt(V):
   Q = []
   for v in V:
       for q in Q:
           v = v - np.dot(v, q) * q
       v = v / np.linalg.norm(v)
       Q.append(v)
   return np.array(Q)
```

```python
V = np.array([[1, 1, 0], [1, 0, 1], [0, 1, 1]])
Q = gram_schmidt(V)
```

QR - Разложение

```python
A = np.array([[1, 1, 0], [1, 0, 1], [0, 1, 1]])
Q, R = np.linalg.qr(A)
```
