```python
import numpy as np
from numpy import linalg as LA
```

```python
a = np.array([3, 5])
V = np.array([
    [1, 9], 
    [2, 0],
    [1, 7]
    ])

np.dot(a, V.T)
```

```python
V / LA.norm(V, axis=1, keepdims=True)
```

```python
def closest_to_antigrad(grad, vectors): 
    sim = np.dot(grad, vectors.T)
    return sim.argmin()
```

```python
closest_to_antigrad(a, V)
```
