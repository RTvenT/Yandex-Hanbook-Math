# __Типы графов__

Лемма о рукопожатиях 

```python
import networkx as nx
import random
```

```python
graph = nx.erdos_renyi_graph(10, 0.3) # Случайный граф с 10 вершинами и вероятностью ребра 0.3

sum_degrees = sum(dict(graph.degree()).values()) # Сумма степеней
double_edges = 2 * graph.number_of_edges()

print(f"Сумма степеней: {sum_degrees}")
print(f"Удвоенное количество рёбер: {double_edges}") 

nx.draw(graph, with_labels=True, node_size=500, node_color='cornflowerblue')
```
