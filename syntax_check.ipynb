```python
import numpy as np
from numpy import linalg as LA
```

```python
a = np.array([3, 5])
V = np.array([
    [1, 9], 
    [2, 0],
    [1, 7]
    ])

np.dot(a, V.T)
```

```python
V / LA.norm(V, axis=1, keepdims=True)
```

```python
def closest_to_antigrad(grad, vectors): 
    sim = np.dot(grad, vectors.T)
    return sim.argmin()
```

```python
closest_to_antigrad(a, V)
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
n, d = map(int, input().split())
X = np.array([list(map(float, input().split())) for _ in range(n)])
y = np.array(list(map(float, input().split())))
a_r, a_l = map(float, input().split())
```

```python
ridge_model = Ridge(fit_intercept=False, alpha=a_r)
lasso_model = Lasso(fit_intercept=False, alpha=a_l)
```

```python
ridge_model.fit(X, y)
lasso_model.fit(X, y)
```

```python
print(*ridge_model.coef_.round(4))
```
